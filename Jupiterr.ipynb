```python
import sys
sys.path.append('/Users/eshaibekuulu/Downloads/Python3/lib/python3.9/site-packages')
import cv2
import numpy as np

def detect_faces_from_camera():
    # Load the pre-trained face detection model
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Open the default camera (0) or a specific camera (e.g., 1)
    video_capture = cv2.VideoCapture(0)

    firstTime = True

    while True:
        # Capture frame-by-frame
        ret, frame = video_capture.read()

        if not ret:
            print("Error: Could not receive frame from camera.")
            break

        if firstTime:
            print(frame, end =" ")
            firstTime = False

        # Convert the frame to grayscale for face detection
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Detect faces in the frame
        faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # Draw rectangles around the detected faces
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

        # Display the resulting frame
        cv2.imshow('Camera', frame)

        # Check for the 'q' key to quit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the video capture object and close all windows
    video_capture.release()
    cv2.destroyAllWindows()

# Call the function to detect faces from the camera feed
detect_faces_from_camera()
```
